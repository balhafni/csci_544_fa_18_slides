```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticks
```

```python
import csv
```

```python
x = []
y = []
count = 0
with open('/Users/alhafni/Desktop/plot_test.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter =',')
    for row in plots:
        if count == 21:
            break
        print(row)
        if row[1]:
            x.append(row[0])
            y.append(float(row[1]))
            count +=1
fig = plt.figure(figsize=(10, 2))
ax = fig.add_subplot(111) 
plt.plot(x,y)
# fig, ax = plt.subplots()
plt.xlabel('Classes')
plt.ylabel('F1-Score')
# plt.legend
# plt.tight_layout()
fig.autofmt_xdate()
plt.show()
```
